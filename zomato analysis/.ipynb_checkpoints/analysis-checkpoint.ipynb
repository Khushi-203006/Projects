```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("Zomato-data-.csv")
print(data) #printing data
data.info() #printing data information

# converting the data type of rate to float by removing the denominator
def handlerate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)

data['rate'] = data['rate'].apply(handlerate) #manipulate data in
print(data)
print(data.info())
```

```python
obj_1= "Objective 1. Determine the most common restaurants in the dataset."
print(obj_1)

sns.countplot(data = data, x= 'listed_in(type)')

plt.xlabel("Type of Restaurant")
plt.show()

print("""Analysis summary:-
      'Dining' is the most frequently occurring restaurant type.
      """)
```

```python
obj_2= "Objective 2. Analyse the distribution of online ordering in different restaurants."
print(obj_2+"\n")

# Define colors
color = ['orange', 'green']

# Creating a count plot for analysis
sns.countplot( data = data, x = 'listed_in(type)',hue = 'online_order' , palette = color)

# Label x-axis
plt.xlabel("Type of restaurant")
plt.show()

print("""Analysis summary:-
      Online Ordering trends are as follows:-
          -> The majority of Dining restaurants do not offer online ordering.
          -> Cafes are likely to support online ordering.
          -> Buffet and Others have low count but do offer online ordering.""")
```

```python
print("Objective 3. To identify the maximum number of votes.Also find Restaurant(s) with maximum votes.\n")
df = pd.DataFrame(data)

#grouping restaurents in votes 
grouped_data = data.groupby('listed_in(type)')['votes'].sum()
vote_count = pd.DataFrame({'votes' : grouped_data})
plt.plot(vote_count ,c="purple" , marker = "*")
plt.xlabel("Types of restrurent" , c = "purple", size=15)
plt.ylabel("Votes",c = 'purple' , size=15)
plt.show()

print("""Analysis summary:-
        Dining got maximum vote over 20000.
        """)

max_votes = data["votes"].max()
rest_max_votes = data.loc[data["votes"] == max_votes , "name"]
max_vote_reataurent = rest_max_votes.tolist()

#tolist() - convert numpy array or pandas series into a regular python list.

print(max_vote_reataurent[0] , "have maximum votes which are", max_votes)

#why [0] = to print element of list instead of whole list
```

```python
print("Objective 4:- Analyse how many restaurents are with online ordering.")

sns.countplot(data = data, x = "online_order")
plt.show()

print("""
        Maximum restaurants donot suggest online order.
        """)
```

```python
print("Objective 5:- To determine the most common rating range among the restaurents in the dataset. \n")
# print(data)
# plt.hist(data = data, x = data["rate"], bins =5)
# plt.show()

def handlerate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)

data['rate'] = data['rate'].apply(handlerate)

sns.histplot(data = data, x = 'rate', hue = 'online_order',multiple='stack',
             bins = np.linspace(2.5,5.0,6))

plt.xlabel("Rating")
plt.ylabel("No.of restaurents")
plt.title("Rating of different restaurents")
plt.show()

sns.histplot(data=data, x = 'rate', hue ='listed_in(type)', multiple = "dodge",
             bins = np.linspace(2.5,5.0,6))
plt.xlabel("Rating")
plt.ylabel("No.of restaurents")
plt.title("Rating of different restaurents")
plt.show()

print("Analysis Summary:- The majority of restaurants received rating ranging from 3.5 to 4")
      
```
